```python
ham = df[df['v1']=='ham']
spam = df[df['v1']=='spam']
```

```python
data = ham.append(spam, ignore_index=True)
data.shape
```

# Data Cleaning part

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(0, len(data)):
    review = re.sub('[^a-zA-Z]',' ', data['v2'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
x
```

```python
data.head()
```

```python
data['v1'] = data['v1'].astype('category')
data['v1'] = data['v1'].cat.codes
```

```python
data.shape
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, data['v1'], test_size=0.25, random_state=1)
```

# Building Niave Bayes Theorem

```python
from sklearn.naive_bayes import MultinomialNB
nbmodel = MultinomialNB().fit(x_train, y_train)
```

```python
y_pred_train = nbmodel.predict(x_train)
y_pred_test = nbmodel.predict(x_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print(confusion_matrix(y_train, y_pred_train))
print()
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_train, y_pred_train))
print()
print(classification_report(y_test, y_pred_test))
```

```python
print(accuracy_score(y_train, y_pred_train))
print()
print(accuracy_score(y_test, y_pred_test))
```
